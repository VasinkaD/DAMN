```python
import numpy as np
import matplotlib.pyplot as plt
import zipfile
```

```python
#Defining the percentiles for the upper and lower bound of errorbars
lower_percentile_border = 5
upper_percentile_border = 95

#Were you using all data for Richardson-Lucy algorithm?
use_all_data_sim = False
use_all_data_exp = False
```

```python
#Deep-STORM
DS_number_of_graph_points = 8
DS_number_of_samples = 1000
```

## Visualize dataset of the Exp graph

### Prepare simulated data visualization

```python
#Loading the relevant data files
horizontal_axis_sim = np.load("Data_files/Generated_data/Simulated_axis_array.npy")
if use_all_data_sim:
    RL_horizontal_axis_sim = horizontal_axis_sim
else:
    RL_horizontal_axis_sim = horizontal_axis_sim[::4]

RL_errors_sim = np.load("Data_files/Processed_data/Sim_RL_errors.npy")
RL_TV_errors_sim = np.load("Data_files/Processed_data/Sim_RLTV_errors.npy")
DAMN_model_errors_sim = np.load("Data_files/Processed_data/Sim_DAMN_model_errors.npy")

DS_horizontal_axis_sim = np.array([50, 70, 97, 135, 187, 259, 360, 500])
DS_models_errors_sim = np.zeros([DS_number_of_graph_points, DS_number_of_samples])
with zipfile.ZipFile("Optical_experiment/DeepStorm_exp_MAEs.zip") as z:
    for i in range(8):
        with z.open("DS_MAEs_array_Experiment_position_" + str(i) + ".npy") as f:
            DS_models_errors_sim[i] = np.load(f)
```

```python
#Process the evaluated per-sample errors to get average value and error bars
RL_averages_sim = np.mean(RL_errors_sim, axis=-1)
RL_low_per_sim = np.percentile(RL_errors_sim, lower_percentile_border, axis=-1)
RL_upp_per_sim = np.percentile(RL_errors_sim, upper_percentile_border, axis=-1)

RL_TV_averages_sim = np.mean(RL_TV_errors_sim, axis=-1)
RL_TV_low_per_sim = np.percentile(RL_TV_errors_sim, lower_percentile_border, axis=-1)
RL_TV_upp_per_sim = np.percentile(RL_TV_errors_sim, upper_percentile_border, axis=-1)

DS_averages_sim = np.mean(DS_models_errors_sim, axis=-1)
DS_low_per_sim = np.percentile(DS_models_errors_sim, lower_percentile_border, axis=-1)
DS_upp_per_sim = np.percentile(DS_models_errors_sim, upper_percentile_border, axis=-1)

DAMN_averages_sim = np.mean(DAMN_model_errors_sim, axis=-1)
DAMN_low_per_sim = np.percentile(DAMN_model_errors_sim, lower_percentile_border, axis=-1)
DAMN_upp_per_sim = np.percentile(DAMN_model_errors_sim, upper_percentile_border, axis=-1)
```

### Prepare measured data visualization

```python
#Loading the relevant data files
horizontal_axis_exp = np.array([50, 70, 97, 135, 187, 259, 360, 500])
if use_all_data_exp:
    RL_horizontal_axis_exp = horizontal_axis_exp
else:
    RL_horizontal_axis_exp = horizontal_axis_exp[::4]

RL_errors_exp = np.load("Data_files/Processed_data/Exp_RL_errors.npy")
RL_TV_errors_exp = np.load("Data_files/Processed_data/Exp_RLTV_errors.npy")
DAMN_model_errors_exp = np.load("Data_files/Processed_data/Exp_DAMN_model_errors.npy")

DS_horizontal_axis_exp = np.array([50, 70, 97, 135, 187, 259, 360, 500])
DS_models_errors_exp = np.zeros([DS_number_of_graph_points, DS_number_of_samples])
with zipfile.ZipFile("Optical_experiment/DeepStorm_exp_MAEs.zip") as z:
    for i in range(8):
        with z.open("DS_MAEs_array_real_Experiment_position_" + str(i) + ".npy") as f:
            DS_models_errors_exp[i] = np.load(f)
```

```python
#Process the evaluated per-sample errors to get average value and error bars
RL_averages_exp = np.mean(RL_errors_exp, axis=-1)
RL_low_per_exp = np.percentile(RL_errors_exp, lower_percentile_border, axis=-1)
RL_upp_per_exp = np.percentile(RL_errors_exp, upper_percentile_border, axis=-1)

RL_TV_averages_exp = np.mean(RL_TV_errors_exp, axis=-1)
RL_TV_low_per_exp = np.percentile(RL_TV_errors_exp, lower_percentile_border, axis=-1)
RL_TV_upp_per_exp = np.percentile(RL_TV_errors_exp, upper_percentile_border, axis=-1)

DS_averages_exp = np.mean(DS_models_errors_exp, axis=-1)
DS_low_per_exp = np.percentile(DS_models_errors_exp, lower_percentile_border, axis=-1)
DS_upp_per_exp = np.percentile(DS_models_errors_exp, upper_percentile_border, axis=-1)

DAMN_averages_exp = np.mean(DAMN_model_errors_exp, axis=-1)
DAMN_low_per_exp = np.percentile(DAMN_model_errors_exp, lower_percentile_border, axis=-1)
DAMN_upp_per_exp = np.percentile(DAMN_model_errors_exp, upper_percentile_border, axis=-1)
```

### Visualize

```python
#Visualize graph
fig, ax = plt.subplots(constrained_layout=True)
fig.set_size_inches(6.5, 5)

#Richardson-Lucy
plt.fill_between(RL_horizontal_axis_sim, RL_low_per_sim, RL_upp_per_sim, color="green", alpha=0.25)
plt.plot(RL_horizontal_axis_sim, RL_averages_sim, color="green", label="RL simulation")

plt.errorbar(RL_horizontal_axis_exp, RL_averages_exp, 
             yerr=[RL_averages_exp-RL_low_per_exp, RL_upp_per_exp-RL_averages_exp], color="darkgreen", linestyle="None")
plt.scatter(RL_horizontal_axis_exp, RL_averages_exp, color="darkgreen", label="RL experiment", marker="^")

#TV Richardson-Lucy
plt.fill_between(RL_horizontal_axis_sim, RL_TV_low_per_sim, RL_TV_upp_per_sim, color="blue", alpha=0.25)
plt.plot(RL_horizontal_axis_sim, RL_TV_averages_sim, color="blue", label="TV RL simulation")

plt.errorbar(RL_horizontal_axis_exp, RL_TV_averages_exp, 
             yerr=[RL_TV_averages_exp-RL_TV_low_per_exp, RL_TV_upp_per_exp-RL_TV_averages_exp], color="darkblue", linestyle="None")
plt.scatter(RL_horizontal_axis_exp, RL_TV_averages_exp, color="darkblue", label="TV RL experiment", marker="^")

#DAMN model
plt.fill_between(horizontal_axis_sim, DAMN_low_per_sim, DAMN_upp_per_sim, color="red", alpha=0.25)
plt.plot(horizontal_axis_sim, DAMN_averages_sim, color="red", label="DAMN simulation")

plt.errorbar(horizontal_axis_exp, DAMN_averages_exp, 
             yerr=[DAMN_averages_exp-DAMN_low_per_exp, DAMN_upp_per_exp-DAMN_averages_exp], color="darkred", linestyle="None")
plt.scatter(horizontal_axis_exp, DAMN_averages_exp, color="darkred", label="DAMN experiment", marker="^")

#DeepStorm
plt.fill_between(DS_horizontal_axis_sim, DS_low_per_sim, DS_upp_per_sim, color="purple", alpha=0.25)
plt.plot(DS_horizontal_axis_sim, DS_averages_sim, color="purple", linestyle="--")
plt.scatter(DS_horizontal_axis_sim, DS_averages_sim, color="purple", label="DS simulation")

plt.errorbar(DS_horizontal_axis_exp, DS_averages_exp, 
             yerr=[DS_averages_exp-DS_low_per_exp, DS_upp_per_exp-DS_averages_exp], color="purple", linestyle="None")
plt.scatter(DS_horizontal_axis_exp, DS_averages_exp, color="purple", label="DS experiment", marker="^")

#Axis settings
ax.set_xlabel("Emitter concentration")
ax.set_ylabel("Error")
plt.xscale("log")
plt.yscale("log")
plt.legend()

plt.show()
```
