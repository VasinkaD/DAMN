```python
import numpy as np
import matplotlib.pyplot as plt
import zipfile
```

## Set global parameters

```python
#Defining the percentiles for the upper and lower bound of errorbars
lower_percentile_border = 5
upper_percentile_border = 95

#Were you using all data for Richardson-Lucy algorithm?
use_all_data_SNR = False
use_all_data_PSF = False
use_all_data_Conc = False
use_all_data_GtoA = False
```

```python
#Deep-STORM
DS_number_of_graph_points = 8
DS_number_of_samples = 1000
```

## Visualize dataset of the SNR graph - panel (A)

```python
#Loading the relevant data files
horizontal_axis_SNR = np.load("Data_files/Generated_data/SNR_axis_array.npy")
if use_all_data_SNR:
    RL_horizontal_axis_SNR = horizontal_axis_SNR
else:
    RL_horizontal_axis_SNR = horizontal_axis_SNR[::4]

RL_errors_SNR = np.load("Data_files/Processed_data/SNR_RL_errors.npy")
RL_TV_errors_SNR = np.load("Data_files/Processed_data/SNR_RLTV_errors.npy")
DAMN_model_errors_SNR = np.load("Data_files/Processed_data/SNR_DAMN_model_errors.npy")

DS_horizontal_axis_SNR = np.logspace(0, 5, DS_number_of_graph_points)
DS_models_errors_SNR = np.zeros([DS_number_of_graph_points, DS_number_of_samples])
with zipfile.ZipFile("DeepStorm_MAEs.zip") as z:
    for i in range(8):
        with z.open("DS_MAEs_array_SNR_position_" + str(i) + ".npy") as f:
            DS_models_errors_SNR[i] = np.load(f)
```

```python
#Process the evaluated per-sample errors to get average value and error bars
RL_averages_SNR = np.mean(RL_errors_SNR, axis=-1)
RL_low_per_SNR = np.percentile(RL_errors_SNR, lower_percentile_border, axis=-1)
RL_upp_per_SNR = np.percentile(RL_errors_SNR, upper_percentile_border, axis=-1)

RL_TV_averages_SNR = np.mean(RL_TV_errors_SNR, axis=-1)
RL_TV_low_per_SNR = np.percentile(RL_TV_errors_SNR, lower_percentile_border, axis=-1)
RL_TV_upp_per_SNR = np.percentile(RL_TV_errors_SNR, upper_percentile_border, axis=-1)

DS_averages_SNR = np.mean(DS_models_errors_SNR, axis=-1)
DS_low_per_SNR = np.percentile(DS_models_errors_SNR, lower_percentile_border, axis=-1)
DS_upp_per_SNR = np.percentile(DS_models_errors_SNR, upper_percentile_border, axis=-1)

DAMN_averages_SNR = np.mean(DAMN_model_errors_SNR, axis=-1)
DAMN_low_per_SNR = np.percentile(DAMN_model_errors_SNR, lower_percentile_border, axis=-1)
DAMN_upp_per_SNR = np.percentile(DAMN_model_errors_SNR, upper_percentile_border, axis=-1)
```

```python
#Visualize graph
fig, ax = plt.subplots(constrained_layout=True)
fig.set_size_inches(6.5, 5)

#Richardson-Lucy
plt.fill_between(RL_horizontal_axis_SNR, RL_low_per_SNR, RL_upp_per_SNR, color="green", alpha=0.25)
plt.plot(RL_horizontal_axis_SNR, RL_averages_SNR, color="green", label="Richardson-Lucy")

#TV Richardson-Lucy
plt.fill_between(RL_horizontal_axis_SNR, RL_TV_low_per_SNR, RL_TV_upp_per_SNR, color="blue", alpha=0.25)
plt.plot(RL_horizontal_axis_SNR, RL_TV_averages_SNR, color="blue", label="TV Richardson-Lucy")

#Deep-STORM model
plt.fill_between(DS_horizontal_axis_SNR, DS_low_per_SNR, DS_upp_per_SNR, color="purple", alpha=0.25)
plt.plot(DS_horizontal_axis_SNR, DS_averages_SNR, color="purple", label="DS model")

#DAMN model
plt.fill_between(horizontal_axis_SNR, DAMN_low_per_SNR, DAMN_upp_per_SNR, color="red", alpha=0.25)
plt.plot(horizontal_axis_SNR, DAMN_averages_SNR, color="red", label="DAMN model")

#Axis settings
def Forward(SNR_):
    Guass_kernel_peak = 0.0795779                 #Value of normed Gaussian kernel with width=2 at the center
    PNR = SNR_ * Guass_kernel_peak
    return PNR

def Inverse(PNR_):
    Guass_kernel_peak = 0.0795779                 #Value of normed Gaussian kernel with width=2 at the center
    SNR = PNR_ / Guass_kernel_peak
    return SNR

ax.set_xlabel("SNR")
ax.set_ylabel("Error")
secax = ax.secondary_xaxis('top', functions=(Forward, Inverse))
secax.set_xlabel("PNR")
plt.xscale("log")
plt.yscale("log")

plt.legend()
plt.show()
```
