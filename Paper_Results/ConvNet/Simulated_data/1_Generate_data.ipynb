```python
import os

import numpy as np
np.random.seed(18)
import matplotlib.pyplot as plt

from func_file_Simulate import *
```

```python
#Create folder for storing generated data files
if not os.path.exists("Data_files"):
    os.mkdir("Data_files")
if not os.path.exists("Data_files/Generated_data"):
    os.mkdir("Data_files/Generated_data")
```

# Generating the data

```python
#The fixed parameters
number_of_samples = 100        #Number of samples per graph point (default 100)
number_of_graph_points = 51    #Number of points on the horizontal axis of each graph (default 51)

base_size = 50                 #Base image size (default 50)
upsampling_factor = 1          #Target model upsampling - No upsamling in this analysis (default 1)
image_size = base_size * upsampling_factor

Poisson_noise = 10             #Average shor noise intensity (default 10)

#These parameters are replaced by horizontal axis values in appropriate places
emitter_power = 5000           #Average emitter power (default 5000)
PSF_width = 2                  #Point spread function width, sigma parameter (default 2)
concentration = 50             #Number of emitters in the base_size image (default 50)
```

## Generate dataset for the SNR graph - panel (A)

```python
#Horizontal axis
SNR_axis_LB = 0    #[10**.] Lower bound of the horizontal axis for the SNR graph (default 0)
SNR_axis_UB = 5    #[10**.] Upper bound of the horizontal axis for the SNR graph (default 5)

SNR_axis_array = np.logspace(SNR_axis_LB, SNR_axis_UB, number_of_graph_points)   #Horizontal axis expressed as Signal-to-noise ratio (= Emitter power / Average noise intensity)
EP_axis_array = SNR_axis_array * Poisson_noise                                   #Horizontal axis expressed as Emitter power
```

```python
#Generating the SNR dataset
SNR_data_low_res = np.zeros((number_of_graph_points, number_of_samples, image_size, image_size))
SNR_data_high_res = np.zeros((number_of_graph_points, number_of_samples, image_size, image_size))

for i in range(number_of_graph_points):
    SNR_data_low_res[i], SNR_data_high_res[i] = Generate_normed_dataset(number_of_samples, base_size, upsampling_factor, 
                                                                        EP_axis_array[i], Poisson_noise, PSF_width, concentration)
```

```python
#Save the dataset
np.save("Data_files/Generated_data/SNR_data_low_res.npy", SNR_data_low_res)
np.save("Data_files/Generated_data/SNR_data_high_res.npy", SNR_data_high_res)
np.save("Data_files/Generated_data/SNR_axis_array.npy", SNR_axis_array)
```

```python
#Show an example of a generated data pair
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.matshow(SNR_data_low_res[15,0], cmap="gray", fignum=False)
plt.title("Low resolution image")

plt.subplot(122)
plt.matshow(SNR_data_high_res[15,0], cmap="gray", fignum=False)
plt.title("High resolution image")

plt.show()
```

```python
#Show examples for various SNR values
plt.figure(figsize=(20,5))
plt.subplot(141)
plt.matshow(SNR_data_low_res[10,0], cmap="gray", fignum=False)
plt.title("SNR: " + str(SNR_axis_array[10]))
plt.subplot(142)
plt.matshow(SNR_data_low_res[20,0], cmap="gray", fignum=False)
plt.title("SNR: " + str(SNR_axis_array[20]))
plt.subplot(143)
plt.matshow(SNR_data_low_res[30,0], cmap="gray", fignum=False)
plt.title("SNR: " + str(SNR_axis_array[30]))
plt.subplot(144)
plt.matshow(SNR_data_low_res[40,0], cmap="gray", fignum=False)
plt.title("SNR: " + str(SNR_axis_array[40]))

plt.show()
```

## Generate dataset for the PSF width graph - panel (B)

```python
#Horizontal axis
PSF_axis_LB = 0    #[10**.] Lower bound of the horizontal axis for the PSF graph (default 0)
PSF_axis_UB = 1    #[10**.] Upper bound of the horizontal axis for the PSF graph (default 1)

PSF_width_axis_array = np.logspace(PSF_axis_LB, PSF_axis_UB, number_of_graph_points)
```

```python
#Save the dataset
np.save("Data_files/Generated_data/PSF_data_low_res.npy", PSF_data_low_res)
np.save("Data_files/Generated_data/PSF_data_high_res.npy", PSF_data_high_res)
np.save("Data_files/Generated_data/PSF_axis_array.npy", PSF_width_axis_array)
```

## Generate dataset for the Gauss-to-Airy kernel transition graph - panel (D)

```python
#Horizontal axis
Transition_axis_LB = 0.01     #Lower bound of the horizontal axis for the Gauss-to-Airy transition graph (default 50)
Transition_axis_UB = 0.99     #Upper bound of the horizontal axis for the Gauss-to-Airy transition graph (default 750)

Transition_axis_array = np.linspace(Transition_axis_LB, Transition_axis_UB, number_of_graph_points)
```
